```python
import pandas as pd

df = pd.read_csv("data/ecommerce_sales.csv")

print(df.shape)
print(df.columns.tolist())

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Category"].value_counts()
df["Region"].value_counts()
df["Payment Mode"].value_counts()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Order Date"].min(), df["Order Date"].max()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Day"] = df["Order Date"].dt.day
```

```python
df
```

```python
df["Month"] = df["Order Date"].dt.to_period("M").astype(str)
```

```python
df
```

```python
sales_time = df.groupby("Month")["Sales"].sum().reset_index()
sales_time
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(sales_time["Month"], sales_time["Sales"])
plt.xticks(rotation=45)
plt.title("Sales Over Time")
plt.show()
```

```python
profit_time = df.groupby("Month")["Profit"].sum().reset_index()

plt.figure(figsize=(12,5))
plt.plot(profit_time["Month"], profit_time["Profit"])
plt.xticks(rotation=45)
plt.title("Profit Over Time")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index()

plt.figure(figsize=(8,5))
plt.bar(region_sales["Region"], region_sales["Sales"])
plt.title("Sales by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
cat_sales = df.groupby("Category")["Sales"].sum().reset_index()

plt.figure(figsize=(8,5))
plt.bar(cat_sales["Category"], cat_sales["Sales"])
plt.title("Sales by Category")
plt.xticks(rotation=45)
plt.show()
```
